# Get data

```python
import pandas as pd
import numpy as np

# Set display options so we can see entire table
pd.options.display.max_rows = 100
pd.options.display.max_columns = 30

# Create dataframe from excel file and inspect
df = pd.read_excel('stats.xlsx')
display(df)
```

# Clean data

```python
# Assign values in row 0 to column names
df.columns = [x.lower() for x in df.loc[0]]
display(df)
```

```python
# Create dataframe w/ regular and post-season stats
df_reg_season = df.loc[1:30]
df_playoffs = df.loc[35:60]

# Combine the two dataframes
df = df_reg_season.append(df_playoffs)

display(df)
```

```python
# Put stats for each season in one row
def copy_row_below():
    for x in range(1, len(df.index) + 1, 2):
        df.iloc[x-1, 1:] = list(df.iloc[x, :-1])
        
copy_row_below()

display(df)
```

```python
# Add column with regular/post season labels
df.loc[1:30, 'reg/post'] = 'reg season'
df.loc[31:, 'reg/post'] = 'post season'
display(df)
```

```python
# Reset index
df.index = (range(1, len(df.index)+1))
display(df)

# Drop the even numbered index rows
indexes_to_drop = list(range(2, len(df.index) + 1, 2))
df = df.drop(indexes_to_drop)
display(df)
```

# Explore data

```python
# Add fg and ft missed columns
df['fg missed'] = df['fga'] - df['fgm']
df['ft missed'] = df['fta'] - df['ftm']
display(df.info())
```

```python
# Define first and second threepeat
first_threepeat_years = ['1990-91', '1991-92', '1992-93']
second_threepeat_years = ['1995-96', '1996-97', '1997-98']

first_threepeat_bool = df['season'].isin(first_threepeat_years)
second_threepeat_bool = df['season'].isin(second_threepeat_years)

df_first_threepeat = df[first_threepeat_bool].copy()
df_second_threepeat = df[second_threepeat_bool].copy()

df_first_threepeat['threepeat'] = 1
df_second_threepeat['threepeat'] = 2

display(df_first_threepeat)
display(df_second_threepeat)
```

```python
# Combine threepeat dataframes
df = df_first_threepeat.append(df_second_threepeat)
display(df)
```

```python
# Change datatypes of columns
df.loc[:,'age':'pf'] = df.loc[:,'age':'pf'].apply(pd.to_numeric)
df.loc[:,'fg missed':'threepeat'] = df.loc[:,'fg missed':'threepeat'].apply(pd.to_numeric)
display(df.info())
```

(PTS + REB + AST + STL + BLK − Missed FG − Missed FT - TO) / GP [1]

```python
# Add columns for efficiency calculations
df['eff_plus'] = df['pts'] + df['reb'] + df['ast'] + df['stl'] + df['blk'] 
df['eff_minus'] = df['fg missed'] + df['ft missed'] + df['tov']
df['eff'] = df['eff_plus'] - df['eff_minus']

# Inspect efficiency per season
display(df[['season','eff', 'reg/post']])
```

```python
# Filter regular season stats and group by threepeat
filt_reg = df['reg/post']=='reg season'
df_reg = df[filt_reg].groupby(by = 'threepeat').mean()
display(df_reg)
```

# Visualize data

```python
from matplotlib import pyplot as plt

x = ['threepeat 1', 'threepeat 2']
y = df_reg['eff'].values 

plt.bar(x, y, color = 'green')
plt.title('MJ efficiency rating between threepeat runs in regular season')
plt.ylabel('efficiency rating')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
filt_post = df['reg/post']=='post season'
df_post = df[filt_post].groupby(by = 'threepeat').mean()
display(df_post)
```

```python
x = ['threepeat 1', 'threepeat 2']
y = df_post['eff'].values

plt.bar(x, y, color = 'green')
plt.title('MJ efficiency rating between threepeat runs in post season')
plt.ylabel('efficiency rating')
plt.style.use('fivethirtyeight')
plt.show()
```
